```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some random data
x = np.random.normal(size=100)
y = 2 * x + np.random.normal(size=100)
```

```python
# Calculate variance and covariance
var_x = np.var(x)
var_y = np.var(y)
cov_xy = np.cov(x, y)[0][1]

# Print the results
print("Variance of x: ", var_x)
print("Variance of y: ", var_y)
print("Covariance of x and y: ", cov_xy)
```

  Covariance Matrix 
    | Var(x)  Cov(x,y) |
    | Cov(x,y) Var(y)  |

    | 0.9689520094770186  1.8539757175585694 |
    | 1.8539757175585694  4.427005898743486  |

```python
# Visualize the data and covariance
plt.scatter(x, y)
plt.arrow(np.mean(x), np.mean(y), np.sqrt(var_x), np.sqrt(var_y), head_width=0.1, head_length=0.1, color='r')
plt.show()
```
