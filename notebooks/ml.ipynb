#### Train - Test - Validation

In machine learning, it is common to split a dataset into training, testing, and validation sets. The training set is used to fit the model, the testing set is used to evaluate the generalization performance of the model, and the validation set is used to tune the hyperparameters of the model.

The training set is a subset of the data used to fit the model. It is used to learn the parameters of the model, such as the coefficients of a linear regression or the weights of a neural network. The model is trained on the training set using an optimization algorithm, such as gradient descent.

The testing set is a separate subset of the data used to evaluate the generalization performance of the model. After the model has been trained on the training set, it is evaluated on the testing set to see how well it performs on unseen data. The testing set is used to estimate the performance of the model on the population and to compare different models.

The validation set is another separate subset of the data used to tune the hyperparameters of the model. The hyperparameters are the parameters of the model that cannot be learned from the data and need to be set by the practitioner. The validation set is used to evaluate the performance of different combinations of hyperparameters and to choose the best combination. The model is trained on the training set and the hyperparameters are chosen based on the performance on the validation set. The final performance of the model is then evaluated on the testing set

$$ \text{logit}(p) = \ln\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p $$

where $\text{logit}(p)$ is the natural logarithm of the odds of the dependent variable $y$ taking the value 1, given the predictor variables $x_1, x_2, \ldots, x_p$, and $\beta_0, \beta_1, \beta_2, \ldots, \beta_p$ are the coefficients (slopes) of the regression model.

$$ p = \frac{1}{1 + e^{-z}} $$

In this formula, $p$ represents the probability of a binary outcome (such as a customer making a purchase or not), and $z$ represents the linear combination of predictor variables in a logistic regression model. The sigmoid function $f(z) = 1 / (1 + e^{-z})$ transforms the linear combination $z$ into a probability $p$ between 0 and 1.

$$ z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p $$

$$ p = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p)}} $$

### Gradient descent

```python
def gradient_descent(X, y, w, alpha, num_iter):
    m = len(y)  # number of samples
    J_history = []  # to store the cost at each iteration

    for i in range(num_iter):
        y_pred = X.dot(w)  # predicted y
        error = y_pred - y  # error
        gradient = (1/m) * X.T.dot(error)  # gradient
        w = w - alpha * gradient  # update w
        J = (1/(2*m)) * np.sum(error**2)  # cost
        J_history.append(J)  # store cost

    return w, J_history
```

### Cost function and parameter update  linear regression ###
$
J(\theta) = \frac{1}{2m} \sum\limits_{i=1}^{m} (h(x^{(i)}, \theta) - y^{(i)})^2
$

The formula for adjusting the model parameters using gradient descent is:

$
\theta_j := \theta_j - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h(x^{(i)}, \theta) - y^{(i)}) x_j^{(i)}
$

### Cost function and parameter update  logistic regression ###
$
J(\theta) = -\frac{1}{m} \sum\limits_{i=1}^{m} [y^{(i)} \log(h(x^{(i)}, \theta)) + (1 - y^{(i)}) \log(1 - h(x^{(i)}, \theta))]
$

$
\theta_j := \theta_j - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h(x^{(i)}, \theta) - y^{(i)}) x_j^{(i)}
$

Let \mathcal{X} be the set of instances, \mathcal{Y} be the set of labels, and \mathbf{x} be an instance of \mathcal{X}. A supervised classification problem can be mathematically formulated as follows:

\begin{align*}
\text{Find } f: \mathcal{X} \to \mathcal{Y} \text{ such that } f(\mathbf{x}) = \text{argmax}_{y \in \mathcal{Y}} P(y | \mathbf{x})
\end{align*}

where $P(y|\mathbf{x})$ is the probability of a label $y$ given an instance $\mathbf{x}$.

```python
# Import libraries
import numpy as np 
from sklearn.cluster import KMeans 
from sklearn import metrics 
import matplotlib.pyplot as plt 

# Load the data
X = np.loadtxt('data.txt')

# Create list of possible values of k
k_values = range(1, 20)

# Initialize empty list to store the errors
errors = []

# Iterate over the possible values of k
for k in k_values:
    # Create a KMeans instance with k clusters
    kmeans = KMeans(n_clusters=k)
    
    # Fit the model to the data
    kmeans.fit(X)
    
    # Compute the error (inertia)
    error = kmeans.inertia_
    
    # Append the error to the list of errors
    errors.append(error)
    
# Plot the errors
plt.plot(k_values, errors)
plt.xlabel('k')
plt.ylabel('Error (inertia)')
```
